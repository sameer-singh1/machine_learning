```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:\ML\data\iris.csv')
```

```python
df
```

```python
 x = df.iloc[:, [0, 1, 2, 3]].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
inertias=[]
for i in range(1, 6):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x)
    inertias.append(kmeans.inertia_)
plt.plot(range(1, 6), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state=4)
y_predict = kmeans.fit_predict(x)
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],s=100,c='blue',label='Cluster1')
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],s=100,c='red',label='Cluster2')
plt.scatter(x[y_predict==2,0],x[y_predict==2,1],s=100,c='green',label='Cluster3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
#plt.xlabel('sepal_length')
#plt.ylabel('sepal_width')
plt.legend()
plt.show()
```
